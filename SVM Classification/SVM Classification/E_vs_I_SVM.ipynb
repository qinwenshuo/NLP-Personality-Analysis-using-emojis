```python
labels = np.hstack(df_data.to_numpy()[:,:1])
data = df_data.to_numpy()[:,1:]
print(len(labels))
print(len(data))
```

```python
LENGTH_DATA = len(labels)

import random
train_example_nums = int(0.85 * LENGTH_DATA)
val_example_nums = int(0.15 * LENGTH_DATA)

random_ids = [i for i in range(LENGTH_DATA)]
random.Random(4).shuffle(random_ids)

random_train_ids = random_ids[:train_example_nums]
random_validation_ids = random_ids[train_example_nums:]

print(len(random_train_ids), len(random_validation_ids))


X_train = np.array(data[random_train_ids])
Y_train = np.array(labels[random_train_ids])

X_vali = np.array(data[random_validation_ids])
Y_vali = np.array(labels[random_validation_ids])
```

```python
svm = LinearSVC(verbose=0)
svm.fit(X_train, Y_train)
```

```python
Y_pred = svm.predict(X_vali)
```

```python
report = classification_report(Y_vali, Y_pred)
```
