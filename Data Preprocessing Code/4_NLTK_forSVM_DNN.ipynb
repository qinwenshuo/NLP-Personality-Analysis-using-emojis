```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')

import csv
from nltk.corpus import stopwords
import re
```

```python
data_path = '/content/drive/MyDrive/NLP_Workspace/Data/'
with open(data_path + 'emoji_detected.csv','r') as file:
  csvreader = csv.reader(file)
  data = list(csvreader)
```

```python
stoplist = stopwords.words('english')
stoplist.extend(["ever", "one", "do","does","make", "go", "us", "to", "get", 
                 "about", "may", "s", ".", ",", "!", "i", "I", '\"', "?", ";", 
                 "--", "--", "would", "could", "”", "Mr.", "Miss", "Mrs.", 
                 "don’t", "said", "can't", "didn't", "aren't", "I'm", "you're", 
                 "they're", "'s", "(", ")", "...", ":", "'", "+", "n't", "*", "@"
                 ,"[", "]", "'d", "'ve", "..", "<", ">", "=", "+", "-", "_" , "/"
                 ,"'m", "'ll", "^", "%", "......", "...."])
stoplist
```

```python
new_data = []

for line in data:
  post = line[1]
  post = re.sub(r"[0-9]", r"", post)
  # post = re.sub(r':[a-zA-z\d_-]+:', r'<emoji>', post)
  tokens = nltk.word_tokenize(post)
  addme = [t.lower() for t in tokens if t.lower() not in stoplist]
  word_count = len(addme)
  if word_count == 0:
    continue
  new_data.append([line[0], ' '.join(addme), word_count, line[3], line[4], line[5], line[6]])
```

```python
header = ['type', 'text', 'word count', 'label', 300, 301, 302]

with open(data_path + "cleaned_data.csv", 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    writer.writerow(header)
    writer.writerows(new_data[1:])
```
