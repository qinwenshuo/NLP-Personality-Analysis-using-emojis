```python
from google.colab import drive
drive.mount('/content/drive')
import csv
import pandas as pd
import re
data_path = '/content/drive/MyDrive/NLP_Workspace/Data/'
```

```python
new_data = []
with open(data_path + 'reorganized_data.csv','r') as file:
  csvreader = csv.reader(file)
  for line in csvreader:
    line[1] = re.sub("('http|http|ftp|https):\/\/([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:\/~+#-]*[\w@?^=%&\/~+#-])", "", line[1])
    line[1] = line[1].strip()
    if line[1] == '':
      continue
    else:
      line[2] = len(line[1].split())
      new_data.append(line)
  

```

```python
with open(data_path + "urlremoved_data.csv", 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    writer.writerows(new_data)
```
