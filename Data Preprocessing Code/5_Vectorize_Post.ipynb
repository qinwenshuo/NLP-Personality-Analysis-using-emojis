Contains 3 million words and phrases and was fit using 300-dimensional word vectors

```python
EMBEDDING_FILE = '/content/drive/MyDrive/NLP_Workspace/Data Preprocessing Code/GoogleNews-vectors-negative300.bin.gz'
```

```python
from gensim.models import KeyedVectors
big_model = KeyedVectors.load_word2vec_format(EMBEDDING_FILE, binary=True)
```

```python
print(big_model["moments"])
```

```python
vectorized_data = []
labels = []
emoji_features = []

data_path = '/content/drive/MyDrive/NLP_Workspace/Data/'
with open(data_path + 'cleaned_data.csv','r') as file: 
  csvreader = csv.reader(file)
  for post in csvreader:
    content = post[1].split()
    sent_vector = np.zeros(300)
    word_count = 0
    for word in content:
      if word in big_model.vocab:
        sent_vector += big_model[word]
        word_count += 1
    # sent_vector = sent_vector / word_count

    vectorized_data.append(sent_vector.tolist())
    labels.append(post[3])
    emoji_features.append([post[4], post[5], post[6]])
```

```python
df_label = pd.DataFrame(labels[1:])
df_data = pd.DataFrame(vectorized_data[1:])
df_emojis = pd.DataFrame(emoji_features[1:])
```
