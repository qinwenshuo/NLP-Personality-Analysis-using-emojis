```python
from google.colab import drive
drive.mount('/content/drive')
import csv
```

```python
data_path = '/content/drive/MyDrive/NLP_Workspace/Data/'
```

```python
TOTAL_NUMBER_OF_PEOPLE = 8675
POST_NUM = 50
TOTAL_NUM_POSTS = 8675 * 50

new_data = [[[" "],[" "],[0]] for i in range(TOTAL_NUM_POSTS)]


with open(data_path+'mbti_1.csv', "r") as file:
  csvreader = csv.reader(file)
  post_id = 0
  for row in csvreader:
    posts = row[1].split("|||")
    for i in posts:
      new_data[post_id][0] = row[0]
      new_data[post_id][1] = i
      new_data[post_id][2] = len(i.split())
      post_id += 1
```

```python
# Reorganize data into a new csv file
with open(data_path + "reorganized_data.csv", 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    writer.writerows(new_data)
```
