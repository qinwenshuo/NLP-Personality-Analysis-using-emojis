```python
import pandas as pd

col_names = ['type', 'text', 'word_num']
df = pd.read_csv(data_path + "urlremoved_data.csv", names = col_names)
df.head(5)
```

```python
df = df.dropna(how='any',axis=0)
```

```python
df.groupby("type").describe()
```

```python
df_ENFJ = df[df['type']=='ENFJ']
df_ENFP = df[df['type']=='ENFP']
df_ENTJ = df[df['type']=='ENTJ']
df_ENTP = df[df['type']=='ENTP']
df_ESFJ = df[df['type']=='ESFJ']
df_ESFP = df[df['type']=='ESFP']
df_ESTJ = df[df['type']=='ESTJ']
df_ESTP = df[df['type']=='ESTP']
df_INFJ = df[df['type']=='INFJ']
df_INFP = df[df['type']=='INFP']
df_INTJ = df[df['type']=='INTJ']
df_INTP = df[df['type']=='INTP']
df_ISFJ = df[df['type']=='ISFJ']
df_ISFP = df[df['type']=='ISFP']
df_ISTJ = df[df['type']=='ISTJ']
df_ISTP = df[df['type']=='ISTP']

shapes = [(int(df_ENFJ.shape[0]),'ENFJ'),
          (int(df_ENFP.shape[0]), 'ENFP'),
          (int(df_ENTJ.shape[0]), 'ENTJ'),
          (int(df_ENTP.shape[0]), 'ENTP'),
          (int(df_ESFJ.shape[0]), 'ESFJ'),
          (int(df_ESFP.shape[0]), 'ESFP'),
          (int(df_ESTJ.shape[0]), 'ESTJ'),
          (int(df_ESTP.shape[0]), 'ESTP'),
          (int(df_INFJ.shape[0]), 'INFJ'),
          (int(df_INFP.shape[0]), 'INFP'),
          (int(df_INTJ.shape[0]), 'INTJ'),
          (int(df_INTP.shape[0]), 'INTP'),
          (int(df_ISFJ.shape[0]), 'ISFJ'),
          (int(df_ISFP.shape[0]), 'ISFP'),
          (int(df_ISTJ.shape[0]), 'ISTJ'),
          (int(df_ISTP.shape[0]), 'ISTP')
]
shapes
```

```python
shapes.sort()
shapes
```

```python
SAMPLE_NUM = shapes[16 // 2][0]
SAMPLE_NUM
```

```python
def reduction_duplication(df, thres):
  num = df.shape[0]
  if num >= thres:
    return df.sample(thres)
  else:
    missing = thres - num
    duplicate = df

    du_time = missing // num
    remain = missing % num
    while du_time > 0:
      duplicate = pd.concat([duplicate, df])
      du_time -= 1
    duplicate = pd.concat([duplicate, df.sample(remain)])
    return duplicate
```

```python
df_ENFJ = reduction_duplication(df_ENFJ, SAMPLE_NUM)
df_ENFP = reduction_duplication(df_ENFP, SAMPLE_NUM)
df_ENTJ = reduction_duplication(df_ENTJ, SAMPLE_NUM)
df_ENTP = reduction_duplication(df_ENTP, SAMPLE_NUM)
df_ESFJ = reduction_duplication(df_ESFJ, SAMPLE_NUM)
df_ESFP = reduction_duplication(df_ESFP, SAMPLE_NUM)
df_ESTJ = reduction_duplication(df_ESTJ, SAMPLE_NUM)
df_ESTP = reduction_duplication(df_ESTP, SAMPLE_NUM)
df_INFJ = reduction_duplication(df_INFJ, SAMPLE_NUM)
df_INFP = reduction_duplication(df_INFP, SAMPLE_NUM)
df_INTJ = reduction_duplication(df_INTJ, SAMPLE_NUM)
df_INTP = reduction_duplication(df_INTP, SAMPLE_NUM)
df_ISFJ = reduction_duplication(df_ISFJ, SAMPLE_NUM)
df_ISFP = reduction_duplication(df_ISFP, SAMPLE_NUM)
df_ISTJ = reduction_duplication(df_ISTJ, SAMPLE_NUM)
df_ISTP = reduction_duplication(df_ISTP, SAMPLE_NUM)
```

```python
shapes = [(int(df_ENFJ.shape[0]),'ENFJ'),
          (int(df_ENFP.shape[0]), 'ENFP'),
          (int(df_ENTJ.shape[0]), 'ENTJ'),
          (int(df_ENTP.shape[0]), 'ENTP'),
          (int(df_ESFJ.shape[0]), 'ESFJ'),
          (int(df_ESFP.shape[0]), 'ESFP'),
          (int(df_ESTJ.shape[0]), 'ESTJ'),
          (int(df_ESTP.shape[0]), 'ESTP'),
          (int(df_INFJ.shape[0]), 'INFJ'),
          (int(df_INFP.shape[0]), 'INFP'),
          (int(df_INTJ.shape[0]), 'INTJ'),
          (int(df_INTP.shape[0]), 'INTP'),
          (int(df_ISFJ.shape[0]), 'ISFJ'),
          (int(df_ISFP.shape[0]), 'ISFP'),
          (int(df_ISTJ.shape[0]), 'ISTJ'),
          (int(df_ISTP.shape[0]), 'ISTP')
]
shapes
```
